# Actuarial & Health Insurance Concepts (Jargonic Summary)

This notebook illustrates a suite of **actuarial and health-finance concepts** in the domain of medical insurance policy-making. The terminology is intentionally **jargonic** — reflective of actuarial exams, solvency regulations, and policy language — but accompanied by detailed notes so the narrative retains pedagogical value.  

---

## 1. Risk Pooling
- **Principle:** By aggregating heterogeneous exposures into a homogeneous bucket, we achieve *reduced coefficient of variation* of individual losses.  
- **Law of Large Numbers (LLN):**  
  \[
  \lim_{n \to \infty} \overline{X}_n = \mu
  \]  
  where \( \overline{X}_n \) is the average claim per-policyholder.  
- **Implication:** Larger pools ⇒ more stable per‑capita claims; reduced premium volatility.

---

## 2. Adverse Selection
- **Phenomenon:** When insurance is voluntary, the morbidity-weighted population differs from the general population.  
- **Result:** Skewed entry, higher expected cost per entrant.  
- **Cascade:** Premium spiral (“death spiral”). Mitigated via mandates, subsidies, or artificial risk equalization.

---

## 3. Moral Hazard
- **Ex post** (utilization shift after being insured):  
  \[
  U(Q) \;|\; \text{price}=0 \;\;>\;\; U(Q) \;|\; \text{price}>0
  \]  
- **Mechanism:** Coinsurance, copayments, and deductibles act as “skin‑in‑the‑game” instruments to partially internalize marginal consumption decision.

---

## 4. Premium Rating (Technical Pricing)
- **Canonical formula:**  
  \[
  P = \mathbb{E}[\text{Claims}] + \text{Expenses} + \text{Risk Margin} - \text{Investment Income}
  \]  
- **Expense loadings:** operational, distributional, acquisition.  
- **Risk margin:** prudential buffer reflecting volatility, calibrated via confidence interval or solvency framework.

---

## 5. Medical Loss Ratio (MLR)
- **Definition:**  
  \[
  \text{MLR} = \frac{\text{Incurred Claims}}{\text{Earned Premiums}}
  \]  
- **Regulatory Thresholds:** e.g., 80% (individual) or 85% (large group) in ACA.  
- **Policy Mechanism:** MLR floors to enforce fair value provision; rebates if violated.

---

## 6. Reserving (Case + IBNR)
- **Case Reserves:** Reported but unpaid claims.  
- **IBNR (Incurred But Not Reported):** actuarial estimate of lagged losses.  
- **Total Reserve:**  
  \[
  R = R_{\text{case}} + R_{\text{IBNR}}
  \]  
- **Triangle methods (e.g., chain ladder)** applied in practice to extrapolate development.

---

## 7. Risk Adjustment
- **Purpose:** Neutralize insurer incentives to cherry‑pick risks.  
- **Mechanism:** Transfer payment based on relative risk score vs market mean.  
- **Risk score definition:** multiplicative morbidity factor:  
  \[
  C_i = R_i \times \text{Base Cost}
  \]  
  where \(R_i\) > 1 indicates above‑average morbidity load.

---

## 8. Reinsurance / Catastrophic Coverage
- **Stop-loss structure:** reinsurer indemnifies portion of loss beyond retention.  
- **Layer notation:**  
  \[
  L \sim \min(X, M) + \max(X-M,0)
  \]  
  where \(X\) is claim, \(M\) threshold.  
- **Policy view:** Stabilizes insurer solvency, crowd‑in effect on participation.

---

## 9. Medical Trend & Health Inflation
- **Trend factor \(t\):**  
  \[
  C_{y} = C_{0}(1+t)^y
  \]  
- **Observation:** Medical \(t\) typically exceeds CPI.  
- **Drivers:** technology, unit cost inflation, demographic pressures.

---

## 10. Credibility Theory
- **Blended estimator:**  
  \[
  \widehat{\mu} = Z \bar{X}_{\text{own}} + (1-Z)\mu_{\text{manual}}
  \]  
- **Z (credibility factor):** function of sample variance and exposure count.  
- **Use case:** smaller blocks assigned lower statistical weight.

---

## 11. Solvency / Capital Adequacy
- **Risk-based capital (RBC):** threshold derived from distribution quantiles.  
- **Stress framework:**  
  \[
  \text{Capital Buffer} = \text{VaR}_{99.5\%}(L) - \mathbb{E}[L]
  \]  
- **Supervisory regimes:** Solvency II, RBC/NAIC, ICS.

---

## 12. Community Rating vs Experience Rating
- **Community:** homogeneous premium across heterogeneity. Equity ↑, fairness ↓.  
- **Experience:** rating derived from observed claims or group morbidity. Fairness ↑, access ↓.

---

## 13. Actuarial Value (AV)
- **Metric:** proportion of expected costs paid by plan.  
- **Formally:**  
  \[
  AV = \frac{\mathbb{E}[\text{Insurer Liability}]}{\mathbb{E}[\text{Total Allowed}}]
  \]

---

## 14. Underwriting Restrictions
- **Traditional:** pre‑existing exclusion, risk-based denials.  
- **Modern policy:** guaranteed issue, adjusted via subsidies + risk adjustment.

---

## 15. Utilization Management
- **Ex ante control:** prior authorization, formularies.  
- **Economic effect:** modifies demand distribution; reduces moral hazard excess.

---

## 16. Provider Remuneration Models
- **Fee-for-service (FFS):** supply‑induced demand risk.  
- **Capitation:** fixed PMPM, risk shifted to provider.  
- **Bundled episode:** encourages efficiency at episode level.

---

## 17. Demographic / Epidemiological Trends
- Aging → mortality / morbidity drift.  
- Incidence/prevalence metrics integrated into long‑term projection models.

---

## 18. Pool Fragmentation
- **Variance-to-mean ratio** scales inversely with pool size.  
- Fragmented structure ⇒ volatility premiums.

---

## 19. Risk Corridors
- **Settlement logic:** Government shares losses/gains beyond tolerance band.  
- **Formulaic:**  
  \[
  S = \alpha \max\{0,(L - (1+\delta)T)\} - \beta \max\{0,((1-\delta)T - L)\}
  \]  
  where \(L\)=losses, \(T\)=target premium.

---

## 20. Stop-Loss for Beneficiaries
- **Out-of-pocket maximum (OOPM):** cap member liability.  
- **Mechanics:**  
  \[
  \text{Patient Liability} = \min(C, \text{OOPM})
  \]  

---

## 21–33 (Selected Advanced Metrics up to Product Mix Risk)
- **Scenario Testing:** deterministic and stochastic shocks to morbidity trend.  
- **Catastrophe/Extreme Risk:** heavy-tail models (Pareto/lognormal).  
- **Expense Allocation:** per‑member vs percentage loads.  
- **Profit Margins:** ratio of underwriting result to premium.  
- **Combined Ratio:**  
  \[
  CR = \text{Loss Ratio} + \text{Expense Ratio}
  \]  
- **Tail Risk:** VaR/TVaR at confidence level.  
- **Claims Development:** loss triangles, chain ladder factors.  
- **Mortality vs Morbidity Tables:** parametric hazard functions.  
- **Chronic Burden:** distributional heterogeneity via indicator risks.  
- **Lapse/Churn:** persistency and anti-selection interaction.  
- **Product Mix Risk:**  
  Weighted average liability shifts due to composition:  
  \[
  C = \sum_i w_i c_i
  \]  
  where \(w_i\) are mix weights, \(c_i\) cost levels.

---

```python
# --- Colab Notebook Skeleton ---

# Title
"""
# Actuarial Concepts for Policy-Making in Medical Insurance
This notebook demonstrates ~50 key concepts using synthetic data.
It is designed as both a learning portfolio and a policy toolkit.
"""

# Table of Contents
"""
1. Risk Pooling
2. Adverse Selection
3. Moral Hazard
4. Premium Setting
5. Medical Loss Ratio (MLR)
...
50. Scenario & Stress Testing
"""

# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)  # reproducibility
```

```python
# --- Example 1: Risk Pooling ---

# Let's simulate 10,000 individuals
n_individuals = 10000

# True average annual healthcare cost per individual
mean_cost = 5000
std_dev_cost = 4000  # large variation among individuals

# Simulate claims using a normal distribution (clipped at zero)
claims = np.random.normal(mean_cost, std_dev_cost, n_individuals)
claims = np.clip(claims, 0, None)  # no negative costs

# Let's see what happens with different pool sizes when estimating avg cost
pool_sizes = [10, 50, 100, 1000, 10000]
sample_means = []

for size in pool_sizes:
    # simulate 1000 samples of this pool size
    sample_means.append([
        np.mean(np.random.choice(claims, size=size, replace=True))
        for _ in range(1000)
    ])

# Plot distribution of sample means
plt.figure(figsize=(10,6))
for i, size in enumerate(pool_sizes):
    sns.kdeplot(sample_means[i], label=f"Pool size {size}")

plt.axvline(mean_cost, color="black", linestyle="--", label="True Mean Cost")
plt.title("Illustration of Risk Pooling Effect")
plt.xlabel("Average Claim Cost per Person")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# --- Example 2: Adverse Selection ---

# Define population parameters
n_population = 10000
share_healthy = 0.7
share_sick = 0.3

# Healthy: lower mean cost
n_healthy = int(n_population * share_healthy)
healthy_claims = np.random.normal(2000, 1000, n_healthy)
healthy_claims = np.clip(healthy_claims, 0, None)

# Sick: higher mean cost
n_sick = n_population - n_healthy
sick_claims = np.random.normal(12000, 5000, n_sick)
sick_claims = np.clip(sick_claims, 0, None)

# Scenario A: Everyone joins (ideal risk pool)
pool_all = np.concatenate([healthy_claims, sick_claims])

# Scenario B: Only sick join (adverse selection)
pool_sick_only = sick_claims

# Scenario C: Mixed, but only 20% of healthy join
pool_mixed = np.concatenate([np.random.choice(healthy_claims, int(0.2*n_healthy)), sick_claims])

# Compare average costs
avg_all = np.mean(pool_all)
avg_sick_only = np.mean(pool_sick_only)
avg_mixed = np.mean(pool_mixed)

print("Average Cost per Person:")
print(f"Everyone joins      : ${avg_all:,.0f}")
print(f"Sick only join      : ${avg_sick_only:,.0f}")
print(f"Mixed pool (sick + limited healthy): ${avg_mixed:,.0f}")
```

```python
# Visualize the per-person costs under each pool
avg_costs = [avg_all, avg_sick_only, avg_mixed]
scenarios = ["Everyone joins", "Sick only", "Mixed (low healthy participation)"]

plt.figure(figsize=(8,5))
sns.barplot(x=scenarios, y=avg_costs, palette="Set2")
plt.ylabel("Average Claim Cost ($)")
plt.title("Impact of Adverse Selection on Premiums")
plt.show()
```

```python
# --- Example 3: Moral Hazard ---

n_individuals = 2000

# Baseline health need (how much care they'd consume if free)
# Let's model it with a lognormal distribution
baseline_need = np.random.lognormal(mean=8, sigma=0.5, size=n_individuals)

# Utility response factors (represent reduced demand when cost-sharing exists)
factor_no_insurance = 0.4   # People consume 40% of baseline if fully out-of-pocket
factor_partial = 0.7        # With co-pay insurance
factor_full = 1.0           # With full coverage

# Actual consumption in each scenario
consumption_no_ins = baseline_need * factor_no_insurance
consumption_partial = baseline_need * factor_partial
consumption_full = baseline_need * factor_full

# Average costs per person
avg_no_ins = np.mean(consumption_no_ins)
avg_partial = np.mean(consumption_partial)
avg_full = np.mean(consumption_full)

print("Average Cost per Person by Coverage Scenario:")
print(f"No insurance   : ${avg_no_ins:,.0f}")
print(f"Partial insurance  : ${avg_partial:,.0f}")
print(f"Full insurance : ${avg_full:,.0f}")
```

```python
# Plot histogram of average spending
scenarios = ["No insurance", "Partial insurance", "Full insurance"]
avg_costs = [avg_no_ins, avg_partial, avg_full]

plt.figure(figsize=(8,5))
sns.barplot(x=scenarios, y=avg_costs, palette="coolwarm")
plt.ylabel("Average Healthcare Spending ($)")
plt.title("Illustration of Moral Hazard")
plt.show()
```

```python
# --- Example 4: Premium Setting ---

n_members = 5000

# Simulated annual claims distribution
claims = np.random.gamma(shape=2.0, scale=3000, size=n_members)

# 1. Expected Claims
expected_claims = np.mean(claims)

# 2. Expenses per member (admin/marketing etc.)
expense_load = 500

# 3. Risk Margin (% of expected claims)
risk_margin_pct = 0.10
risk_margin = expected_claims * risk_margin_pct

# 4. Investment Income Offset (e.g., insurer earns 2% on reserves)
investment_income = expected_claims * 0.02

# Premium calculation
gross_premium = expected_claims + expense_load + risk_margin - investment_income

print("Premium Calculation Breakdown (per member per year):")
print(f"Expected Claims     : ${expected_claims:,.0f}")
print(f"Expense Load        : ${expense_load:,.0f}")
print(f"Risk Margin (10%)   : ${risk_margin:,.0f}")
print(f"Investment Income   : -${investment_income:,.0f}")
print("-"*40)
print(f"Required Premium    : ${gross_premium:,.0f}")
```

```python
# Explore how premium changes with different risk margins
risk_margins = np.arange(0, 0.21, 0.05)
premiums = [
    expected_claims + expense_load + expected_claims*rm - investment_income
    for rm in risk_margins
]

plt.figure(figsize=(8,5))
sns.lineplot(x=risk_margins*100, y=premiums, marker="o")
plt.xlabel("Risk Margin (%)")
plt.ylabel("Premium per Member ($)")
plt.title("Premium Sensitivity to Risk Margin")
plt.grid(True)
plt.show()
```

```python
# --- Example 5: Medical Loss Ratio ---

n_insurers = 3
np.random.seed(123)

# Simulate total premium collected and total claims paid for each insurer
premiums = np.array([50_000_000, 70_000_000, 100_000_000])
claims   = np.array([42_000_000, 50_000_000, 78_000_000])

# Calculate MLR = Claims / Premium
mlr = claims / premiums

# Put into DataFrame for clarity
insurers = pd.DataFrame({
    "Insurer": ["A", "B", "C"],
    "Premiums": premiums,
    "Claims": claims,
    "MLR": mlr
})

print("MLR by Insurer:\n")
print(insurers.round(3))
```

```python
# Plot MLR as bar chart
plt.figure(figsize=(8,5))
sns.barplot(x="Insurer", y="MLR", data=insurers, palette="viridis")

# Add regulatory threshold line (e.g. 80%)
plt.axhline(0.8, color="red", linestyle="--", label="Regulatory Minimum (80%)")

plt.title("Medical Loss Ratio by Insurer")
plt.ylabel("MLR (Claims / Premiums)")
plt.ylim(0,1)
plt.legend()
plt.show()
```

```python
# --- Example 6: Reserving (IBNR) ---

n_claims = 2000

# Simulate true incurred claim amounts
true_claims = np.random.gamma(shape=2, scale=4000, size=n_claims)

# Reporting lag: some claims reported now, others will surface later
report_prob = 0.7  # 70% reported this year
is_reported = np.random.binomial(1, report_prob, size=n_claims)

reported_claims = true_claims[is_reported == 1]
unreported_claims = true_claims[is_reported == 0]

# Case reserve (reported but unpaid)
case_reserves = np.sum(reported_claims)

# IBNR (estimated for unreported claims)
ibnr_estimate = np.mean(unreported_claims) * len(unreported_claims)

# Total reserve requirement
total_reserve = case_reserves + ibnr_estimate

print("Reserving Illustration:")
print(f"Reported Claims (Case Reserve) : ${case_reserves:,.0f}")
print(f"IBNR Estimate                  : ${ibnr_estimate:,.0f}")
print("-"*40)
print(f"Total Reserve Requirement       : ${total_reserve:,.0f}")
```

```python
# Visualize reported vs unreported distribution
plt.figure(figsize=(8,5))
sns.histplot(reported_claims, color="blue", bins=30, stat="density", label="Reported Claims", alpha=0.6)
sns.histplot(unreported_claims, color="orange", bins=30, stat="density", label="Unreported Claims", alpha=0.6)

plt.xlabel("Claim Amount ($)")
plt.ylabel("Density")
plt.title("Reported vs Unreported (IBNR) Claims")
plt.legend()
plt.show()
```

```python
# --- Example 7: Risk Adjustment ---

n_members = 5000

# Generate risk scores (Healthy ~0.5–1.0, Sick ~1.5–3.0)
healthy_scores = np.random.uniform(0.5, 1.0, int(n_members*0.6))
sick_scores = np.random.uniform(1.5, 3.0, int(n_members*0.4))

population_scores = np.concatenate([healthy_scores, sick_scores])
np.random.shuffle(population_scores)

# Split between 2 insurers
insurer_A_scores = population_scores[:int(0.5*n_members)]  # A gets more healthy
insurer_B_scores = population_scores[int(0.5*n_members):]  # B gets more sick

# Assume claims = base_cost * risk_score
base_cost = 4000
claims_A = insurer_A_scores * base_cost
claims_B = insurer_B_scores * base_cost

avg_claim_A = np.mean(claims_A)
avg_claim_B = np.mean(claims_B)

print("Average Cost per Member (before risk adjustment):")
print(f"Insurer A (healthier enrollees): ${avg_claim_A:,.0f}")
print(f"Insurer B (sicker enrollees)   : ${avg_claim_B:,.0f}")

# Risk adjustment: transfer based on average risk score relative to market
avg_risk_market = np.mean(population_scores)
avg_risk_A = np.mean(insurer_A_scores)
avg_risk_B = np.mean(insurer_B_scores)

transfer_A = base_cost * (avg_risk_market - avg_risk_A) * len(insurer_A_scores)
transfer_B = base_cost * (avg_risk_market - avg_risk_B) * len(insurer_B_scores)

print("\nRisk Adjustment Transfers (positive=receive, negative=pay):")
print(f"Insurer A: ${transfer_A:,.0f}")
print(f"Insurer B: ${transfer_B:,.0f}")

# Adjusted average claim costs
adj_claim_A = (np.sum(claims_A) + transfer_A) / len(insurer_A_scores)
adj_claim_B = (np.sum(claims_B) + transfer_B) / len(insurer_B_scores)

print("\nAverage Cost per Member (after risk adjustment):")
print(f"Insurer A: ${adj_claim_A:,.0f}")
print(f"Insurer B: ${adj_claim_B:,.0f}")
```

```python
labels = ["Before Adjustment", "After Adjustment"]
costs_A = [avg_claim_A, adj_claim_A]
costs_B = [avg_claim_B, adj_claim_B]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(8,5))
ax.bar(x - width/2, costs_A, width, label="Insurer A (healthier)")
ax.bar(x + width/2, costs_B, width, label="Insurer B (sicker)")

ax.set_ylabel("Average Cost per Member ($)")
ax.set_title("Impact of Risk Adjustment")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

```python
# --- Example 8: Reinsurance & Catastrophic Coverage ---

n_members = 10000

# Claims: many small, few catastrophic
claims = np.random.exponential(scale=5000, size=n_members)

# Add some very large catastrophic claims
claims[np.random.choice(n_members, 20, replace=False)] *= 20

# Reinsurance stop-loss: threshold at 50,000
threshold = 50_000

# Insurer pays up to threshold per member
insurer_costs = np.minimum(claims, threshold)

# Reinsurer covers excess above threshold
reinsurer_costs = np.maximum(claims - threshold, 0)

print("Total Costs Split:")
print(f"Total Claims       : ${np.sum(claims):,.0f}")
print(f"Insurer Retained   : ${np.sum(insurer_costs):,.0f}")
print(f"Reinsurer Covered  : ${np.sum(reinsurer_costs):,.0f}")
print(f"Share Covered by Reinsurer: {np.sum(reinsurer_costs)/np.sum(claims):.1%}")
```

```python
# --- Example 9: Health Trend Projection ---

years = np.arange(2025, 2036)
base_cost = 5000  # current avg annual per-member cost

medical_trend = 0.06   # 6% medical inflation
general_inflation = 0.03  # 3% CPI

medical_costs = base_cost * (1 + medical_trend) ** (years - 2025)
cpi_costs = base_cost * (1 + general_inflation) ** (years - 2025)

plt.figure(figsize=(8,5))
plt.plot(years, medical_costs, label="Medical Cost Trend (6%)")
plt.plot(years, cpi_costs, label="General Inflation (3%)", linestyle="--")
plt.ylabel("Average Annual Cost ($)")
plt.title("Healthcare Cost Trend vs. CPI Inflation")
plt.legend()
plt.show()
```

```python
# --- Example 16: Payment Models ---

visits = np.random.poisson(5, 1000)  # avg 5 visits per member
fee_rate = 200
cap_rate = 1000
bundle_rate = 1500  # per member episode

cost_fee = visits * fee_rate
cost_cap = np.repeat(cap_rate, 1000)
cost_bundle = np.repeat(bundle_rate, 1000)

print("Average insurer payment per member:")
print(f"Fee-for-service: ${np.mean(cost_fee):,.0f}")
print(f"Capitation     : ${np.mean(cost_cap):,.0f}")
print(f"Bundled        : ${np.mean(cost_bundle):,.0f}")
```

```python
# --- Example 18: Pool Fragmentation ---

claims = np.random.gamma(2, 3000, 10000)

def volatility(size):
    sims = [np.mean(np.random.choice(claims, size=size)) for _ in range(1000)]
    return np.std(sims)/np.mean(sims)

sizes = [50, 500, 5000]
for s in sizes:
    print(f"Relative volatility with pool {s}: {volatility(s):.2%}")
```

```python
# --- Example 19: Risk Corridors ---

target = 5000
claims = np.mean(np.random.gamma(2, 3000, 1000))

if claims < 0.97*target:
    settlement = 0.5*(target-claims)  # insurer returns part
elif claims > 1.03*target:
    settlement = -0.5*(claims-target)  # govt reimburses part
else:
    settlement = 0

print(f"Claims: ${claims:,.0f}, Settlement: ${settlement:,.0f}")
```

```python
# --- Example 20: Out-of-Pocket Max ---

claims = np.random.gamma(2, 2000, 1000)
oop_max = 3000
oop_costs = np.minimum(claims, oop_max)

print(f"Avg actual claims: ${np.mean(claims):,.0f}")
print(f"Avg patient OOP with cap: ${np.mean(oop_costs):,.0f}")
```
